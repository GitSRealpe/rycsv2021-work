# **1. Cinemática Robot Movil Kuboki**
---
Caracteristicas chasis del robot:

Este robot posee dos ruedas estandar fijas las cuales restringen el deslizamiento lateral del robot en ambas direcciones, cuenta con dos ruedas de apoyo libres de tracción, las cuales, como su nombre lo dice, sirven de sutentación tanto en la parte delantera como tracera del chasis del robot.

- $l:$ 120 mm - Distancia del centro a cada rueda
- $r:$ 25 mm - Radio de rueda
- $ \alpha_1:$ -90° - Angulo ubicación rueda 1 respecto al  centro
- $ \alpha_2:$ 90° - Angulo ubicación rueda 2 respecto al  centro
- $ \beta_1:$ 180° - Angulo de la rueda 1 respecto al chasi
- $ \beta_2:$ 0° - Angulo de las ruedas respecto al chasis, según su sentido de giro.

Así, con estos datos podemos establecer el modelo cinematico de nuestro Kuboki; al observar su diseño y caracteristicas podemos definirlo como un robot de accionamiento diferencial.

Restricción de rodadura: $ [sin(\alpha + \beta), -cos(\alpha + \beta), (-l)cos(\beta)]R(\theta)\dot{\xi_I} - r\dot{\varphi} = 0$

Restriccion de deslizamiento: $ [cos(\alpha + \beta), sin(\alpha + \beta), lsin(\beta)]R(\theta)\dot{\xi_I} = 0$

Estas restricciones se simplifican bastante gracias a que $\alpha$ y $\beta$ son siempre constantes.

# **3. Modelo Cinemático Robotino**
---
Caracteristicas del chasis:

El Robotino posee tres ruedas suecas de 90° con 6 rodillos, orientadas a 120° una de la otra, cada una con accionamiento independiente. Por lo anterior, la cinematica del robotino no presenta restriccion de deslizamiento, solo de rodadura.

- $l:$ 150 mm - Distancia del centro al plano de cada rueda
- $r:$ 40 mm - Radio de rueda
- $ \alpha_1:$ 60° - Angulo ubicación rueda 1 respecto al  centro
- $ \alpha_2:$ 180° - Angulo ubicación rueda 2 respecto al  centro
- $ \alpha_3:$ 300° - Angulo ubicación rueda 3 respecto al  centro
- $ \beta_{1,2,3}:$ 0° - Angulo de las ruedas respecto al chasis, todos son cero debido a que las 3 ruedas son tangenciales al cuerpo del robot.
- $ \gamma_{1,2,3}:$ 0° - Angulo entre los ejes de los rodillos y el eje principal, igual para todas las ruedas. 

Para este caso las ruedas tampoco son orientavles, por lo tanto $\beta$ tambien es constante.

Con todo lo anterior podemos calcular el modelo del robot basados en la restriccion de rodadura de las 3 ruedas, cuya ecuacion es:

$ [sin(\alpha + \beta + \gamma), -cos(\alpha + \beta + \gamma), (-l)cos(\beta + \gamma)]R(\theta)\dot{\xi_I} - r\dot{\varphi}cos(\gamma) = 0$
